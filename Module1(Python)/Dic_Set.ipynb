```python
#Write a program to create a dictionary from two lists: one of keys and one of values
keys = ["name", "age", "city"]
values = ["Rahul", 21, "Delhi"]

my_dict = dict(zip(keys, values))

print("Dictionary:", my_dict)
```

```python
#Merge two dictionaries into one
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}

merged_dict = {**dict1, **dict2}

print("Merged Dictionary:", merged_dict)
```

```python
#Write a program to sort a dictionary by its values
my_dict = {"apple": 50, "banana": 20, "mango": 40}

sorted_dict = dict(sorted(my_dict.items(), key=lambda item: item[1]))

print("Dictionary sorted by values:")
print(sorted_dict)
```

```python
#Given two sets, check if one set is a subset of another
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}

if set1.issubset(set2):
    print("Set1 is a subset of Set2")
else:
    print("Set1 is not a subset of Set2")
```

```python
#Write a program to check whether two lists have at least one common element using sets
list1 = [1, 2, 3, 4]
list2 = [5, 6, 3, 8]

set1 = set(list1)
set2 = set(list2)

if set1.intersection(set2):
    print("Lists have at least one common element")
else:
    print("No common elements found")
```
