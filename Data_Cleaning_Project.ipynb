# Step 1: Load and Inspect Dataset

In this step, we import the required libraries and load the dataset.


```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Messy_Employment_India_Dataset.csv")

# Display first 5 rows to see first few rows to check what data is available
df.head()
```

# Step 2: Understand Dataset Structure

We will check the shape, columns, and data types of the dataset.
This will help us to identify which columns need cleaning.

```python
# Shape of dataset
print("Rows and Columns:", df.shape)

# Column names
print(df.columns)

# Data types
df.info()

# Statistical summary
df.describe(include="all")
```

# Step 3: Handle Missing Values

Missing values can affect analysis and machine learning models.
We replace missing values with the mode and numerical values with the median.

```python
# Check missing values
df.isnull().sum()
```

# Step 4: Remove Duplicate Records

Duplicate rows can distort analysis.
We identify and remove duplicate records.

```python
print("Duplicates:", df.duplicated().sum())

df = df.drop_duplicates()

print("After Removal:", df.duplicated().sum())
```

# Step 5: Correct Inconsistent Formatting

Text values contain inconsistent capitalization and spacing.
We standardize them for consistency.

# Step 6: Fix Data Types

We convert columns to appropriate data types.

```python
df['Date Recorded'] = pd.to_datetime(
    df['Date Recorded']
)

df.info()
```

# Step 7: Detect and Handle Outliers

Outliers are extreme values that can affect analysis.
We use the IQR method.

# Step 8: Standardize Column Names

Column names should follow a consistent format.

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "")
    .str.replace(")", "")
)

df.columns
```

# Step 9: Validate Cleaned Data

We verify that the dataset is clean and ready for analysis.

```python
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicates:")
print(df.duplicated().sum())

df.head()
```

# Step 10: Export Cleaned Dataset

The cleaned dataset is exported for future analysis.

```python
df.to_csv(
    "Cleaned_Employment_India_Dataset.csv",
    index=False
)

print("Dataset Exported Successfully")
```

# Conclusion

The dataset was successfully cleaned by handling missing values,
removing duplicates, standardizing text values ,
correcting data types, treating outliers ,
and exporting a clean version for analysis.
