Module 2 chapter 2 merge

```python
import pandas as pd
df1=pd.DataFrame({
    "a":[1,2,3,4],
    "b":['A','B','C','D']
})
df2=pd.DataFrame({
    "a":[1,2,3,4],
    "d":['a','b','c','d']
})
pd.merge(df1,df2)
```

```python
import pandas as pd
df1=pd.DataFrame({
    "a":[1,2,3,4],
    "b":['A','B','C','D']
})
df2=pd.DataFrame({
    "a":[1,5,7,3],
    "c":['a','b','c','d']
})
pd.merge(df1,df2,on="a", how='outer', indicator=True)
```

```python
import pandas as pd
df1=pd.DataFrame({
    "a":[1,2,3,4],
    "b":['A','B','C','D']
})
df2=pd.DataFrame({
    "a":[1,5,7,3],
    "c":['a','b','c','d']
})
pd.merge(df1,df2,on="a", how='inner', indicator=True)
```

```python
import pandas as pd
df1=pd.DataFrame({
    "a":[1,2,3,4],
    "b":['A','B','C','D']
})
df2=pd.DataFrame({
    "a":[1,5,7,3],
    "c":['a','b','c','d']
})
pd.merge(df1,df2,on="a", how='left',indicator=True)
```

```python
import pandas as pd
df1=pd.DataFrame({
    "a":[1,2,3,4],
    "b":['A','B','C','D']
})
df2=pd.DataFrame({
    "a":[1,5,7,3],
    "c":['a','b','c','d']
})
pd.merge(df1,df2,on="a", how='right',indicator=True)
```

index

```python
import pandas as pd
df1=pd.DataFrame({
    "a":[1,2,3,4,5],
    "b":['A','B','C','D','E']
})
df2=pd.DataFrame({
    "a":[1,2,3,4,5],
    "b":['a','b','c','d','e']
})
pd.merge(df1,df2,left_index=True,right_index=True)
```

suffix

```python
import pandas as pd
df1=pd.DataFrame({
    "a":[1,2,3,4,5],
    "b":['A','B','C','D','E']
})
df2=pd.DataFrame({
    "a":[1,2,3,4,5],
    "b":['a','b','c','d','e']
})
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=('_id','_name'))
```
